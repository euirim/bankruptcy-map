```python
import spacy
nlp = spacy.load('en_core_web_sm')
```

```python
txt = open("sawmill.txt").read()
print(txt)
```

```python
# Read in entire document
doc = nlp(txt)
print(doc.text)
```

```python
# Tokenized Document
for token in doc:
    print(token.text, token.pos_, token.dep_)
```

```python
# Chunks
for chunk in doc.noun_chunks:
    print(chunk.text)
```

```python
# Entities
for ent in doc.ents:
    if ent.label_ == "ORG":
        print(ent.text, ent.label_)
```
