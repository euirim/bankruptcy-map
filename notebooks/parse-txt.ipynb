```python
import spacy
from spacy import displacy
from blackstone.displacy_palette import ner_displacy_options
nlp = spacy.load('en_core_web_sm')
```

```python
txt = open("../txt-files/pgb38llc.txt").read()
```

```python
# Read in entire document
doc = nlp(txt)
print(doc.text)
```

```python
# EXTRACT CHAPTER
# for token in doc:
#     print(token.text, token.pos_, token.dep_)
for i, token in enumerate(doc):
    if token.text.lower() == "chapter" and doc[i+1].pos_ == "SPACE" and doc[i+2].pos_ == "NUM":
            print(doc[i], doc[i+2])
```

```python
# Chunks
for chunk in doc.noun_chunks:
    print(chunk.text)
```

```python
# Entities
for ent in doc.ents:
    if ent.label_ == "ORG":
        print(ent.text, ent.label_)
```

```python
displacy.render(doc, style="ent", options=ner_displacy_options)
```
